```python
import os
import pandas as pd
import numpy as np
import re
import cv2
```

```python
directory = '/home/gaurangajitk/DL/data/sports-video-data/train_images'
sport_dic = {
    'baseball': 0,
    'basketball': 1,
    'boxing': 2,
    'football': 3,
    'volleyball': 4
}
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    temp_keys = re.split(r'(\d+)', text)
    temp_keys.pop(-1)
    try:
        key0 = ''.join(temp_keys[:-1])
        key1 = temp_keys[-1]
    except TypeError:
        print(temp_keys)
        raise TypeError  
    return key0, int(key1)
```

```python
frames = []
labels = []
for sport in os.listdir(directory):
    sport_dir = os.path.join(directory, sport)
    for filename in sorted(os.listdir(sport_dir), key=natural_keys):
        f = os.path.join(sport_dir, filename)
        frames.append(f)
        labels.append(sport_dic[sport])
df = pd.DataFrame(list(zip(frames, labels)), columns=['frame', 'label'])
```

```python
# df.sort_values(by='frame',key=int, inplace=True)
df
```

```python
df['start'] = np.nan
df['end'] = np.nan
df['video_id'] = 0
```

```python
id = 0
for i in range(len(df)-1):
    df.start.iloc[i] = df.frame.iloc[i][-5] == '0'
    df.end.iloc[i] = df.frame.iloc[i+1][-5] == '0'
    if df.start.iloc[i]:
        id += 1
    df.video_id.iloc[i] = id
df.start.iloc[len(df)-1] = False
df.end.iloc[len(df)-1] = True
df.video_id.iloc[len(df)-1] = id
```
